```python
# ============================================================
# 6. KERAS SIMPLE NEURAL NETWORK
# ============================================================
# Keras is a high-level API used to build neural networks easily.
# In TensorFlow 2.x, Keras is included inside TensorFlow.

# Dataset for OR gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    [0],
    [1],
    [1],
    [1]
])

# Create a simple neural network model
model = keras.Sequential([
    layers.Dense(4, activation='relu', input_shape=(2,)),
    layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model
model.fit(X, y, epochs=500, verbose=0)

# Test the model
predictions = model.predict(X)

print("Keras OR Gate Predictions:")
for i in range(len(X)):
    print(X[i], "=>", round(predictions[i][0]))
```

```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

print("PyTorch Version:", torch.__version__)
```

```python
# ============================================================
# 2. CHECK GPU AVAILABILITY
# ============================================================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Using Device:", device)

if torch.cuda.is_available():
    print("GPU Name:", torch.cuda.get_device_name(0))
```

```python
# ============================================================
# 3. BASIC TENSOR OPERATIONS
# ============================================================

a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6], [7, 8]])

print("Tensor A:")
print(a)

print("\nTensor B:")
print(b)

print("\nAddition:")
print(a + b)

print("\nMatrix Multiplication:")
print(torch.matmul(a, b))
```

```python
# ============================================================
# 4. AUTOGRAD EXAMPLE
# ============================================================
# Autograd automatically calculates gradients.
# Gradients are used during backpropagation.

x = torch.tensor(4.0, requires_grad=True)

y = x**2 + 3*x + 5

y.backward()

print("Function value:", y.item())
print("Gradient dy/dx at x = 4:", x.grad.item())
```

```python
# ============================================================
# 5. SIMPLE NEURAL NETWORK FOR OR GATE
# ============================================================

# Input data
X = torch.tensor([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=torch.float32)

# Output labels
y = torch.tensor([
    [0],
    [1],
    [1],
    [1]
], dtype=torch.float32)

# Move data to GPU if available
X = X.to(device)
y = y.to(device)
```

```python
# ============================================================
# 6. DEFINE MODEL
# ============================================================
# nn.Module is the base class for all neural networks in PyTorch.

class ORGateModel(nn.Module):
    def __init__(self):
        super(ORGateModel, self).__init__()

        # First hidden layer: 2 input neurons -> 4 hidden neurons
        self.hidden = nn.Linear(2, 4)

        # Output layer: 4 hidden neurons -> 1 output neuron
        self.output = nn.Linear(4, 1)

        # Activation function
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Forward pass
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x


model = ORGateModel().to(device)

print(model)
```

```python
# ============================================================
# 7. DEFINE LOSS FUNCTION AND OPTIMIZER
# ============================================================

# Binary Cross Entropy is used for binary classification
criterion = nn.BCELoss()

# Adam optimizer updates the model weights
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# ============================================================
# 8. TRAIN THE MODEL
# ============================================================

epochs = 1000

for epoch in range(epochs):

    # Forward pass
    predictions = model(X)

    # Calculate loss
    loss = criterion(predictions, y)

    # Clear old gradients
    optimizer.zero_grad()

    # Backward pass: calculate new gradients
    loss.backward()

    # Update weights
    optimizer.step()

    # Print loss after every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
# ============================================================
# 9. TEST THE MODEL
# ============================================================

with torch.no_grad():
    predictions = model(X)

print("Predictions:")

for i in range(len(X)):
    input_value = X[i].cpu().numpy()
    predicted_value = predictions[i].item()
    final_output = round(predicted_value)

    print(input_value, "=>", final_output)
```
