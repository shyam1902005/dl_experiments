```python
import time
import tensorflow as tf

from tensorflow.keras.datasets import reuters
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense

# Load dataset
vocab_size = 10000
max_length = 200

(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=vocab_size)

# Pad sequences
x_train = pad_sequences(x_train, maxlen=max_length)
x_test = pad_sequences(x_test, maxlen=max_length)

# Function to create Simple RNN model
def create_rnn_model():
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
        SimpleRNN(64),
        Dense(46, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Function to create LSTM model
def create_lstm_model():
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
        LSTM(64),
        Dense(46, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Train and evaluate RNN
rnn_model = create_rnn_model()

start_time = time.time()

rnn_model.fit(
    x_train,
    y_train,
    epochs=3,
    batch_size=64,
    validation_split=0.2
)

rnn_training_time = time.time() - start_time

rnn_loss, rnn_accuracy = rnn_model.evaluate(x_test, y_test)

# Train and evaluate LSTM
lstm_model = create_lstm_model()

start_time = time.time()

lstm_model.fit(
    x_train,
    y_train,
    epochs=3,
    batch_size=64,
    validation_split=0.2
)

lstm_training_time = time.time() - start_time

lstm_loss, lstm_accuracy = lstm_model.evaluate(x_test, y_test)

# Compare results
print("\nPerformance Comparison")
print("----------------------")
print("Simple RNN Accuracy:", rnn_accuracy)
print("Simple RNN Training Time:", rnn_training_time, "seconds")
print("LSTM Accuracy:", lstm_accuracy)
print("LSTM Training Time:", lstm_training_time, "seconds")
```
