```python
def step(z):
    return 1 if z >= 0 else 0


def perceptron(x1, x2, w1, w2, bias):
    z = w1*x1 + w2*x2 + bias
    return step(z)


def AND_gate(x1, x2):
    return perceptron(x1, x2, 1, 1, -1.5)


def OR_gate(x1, x2):
    return perceptron(x1, x2, 1, 1, -0.5)


def NAND_gate(x1, x2):
    return perceptron(x1, x2, -1, -1, 1.5)


def NOR_gate(x1, x2):
    return perceptron(x1, x2, -1, -1, 0.5)


# XOR using multiple perceptrons
def XOR_gate(x1, x2):
    nand_out = NAND_gate(x1, x2)
    or_out = OR_gate(x1, x2)
    return AND_gate(nand_out, or_out)


# XNOR using XOR + NOT
def NOT_gate(x):
    return step(-x + 0.5)


def XNOR_gate(x1, x2):
    xor_out = XOR_gate(x1, x2)
    return NOT_gate(xor_out)


# Testing all gates
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("x1 x2 | AND OR NAND NOR XOR XNOR")
print("--------------------------------")

for x1, x2 in inputs:
    print(
        x1, x2, "|",
        AND_gate(x1, x2),
        OR_gate(x1, x2),
        NAND_gate(x1, x2),
        NOR_gate(x1, x2),
        XOR_gate(x1, x2),
        XNOR_gate(x1, x2)
    )
```
