```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Load built-in IMDB dataset
vocab_size = 10000
max_length = 200

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

# Pad sequences to same length
x_train = pad_sequences(x_train, maxlen=max_length)
x_test = pad_sequences(x_test, maxlen=max_length)

# Build RNN model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    SimpleRNN(64),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Train model
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)

# Evaluate model
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
pred = model.predict(x_test[1001:1002])[0][0]

print("Predicted: ","Positive" if pred >=0.5 else "Negative")
print("Actual: ","Positive" if y_test[1001]==1 else "Negative")
```

```python
# Load original dataset again for showing review text
(_, _), (x_test_original, _) = imdb.load_data()

word_index = imdb.get_word_index()
reverse_word_index = {value + 3: key for key, value in word_index.items()}

def decode_review(review):
    return " ".join(reverse_word_index.get(i, "?") for i in review)

i = 1001

pred = model.predict(x_test[i:i+1])[0][0]

print("Predicted:", "Positive" if pred >= 0.5 else "Negative")
print("Actual:", "Positive" if y_test[i] == 1 else "Negative")
print("Probability:", pred)

print("\nReview:")
print(decode_review(x_test_original[i]))
```
